# GDP Per Capita figure
This code is used to create the GDP per capita figure. Information on the GDP per capita is collected from the [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD?most_recent_value_desc=true)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
File = pd.read_excel('/Users/isamarcortes/Dropbox/Isamar/Papers_In_Prep/Paper_1/SupplementaryMaterial/DatabaseFinalized.xlsx',sheet_name='GDP_PerCapita')
print(File)

```

```python
plt.bar(File['Country'],File['GDP Per Capita'])
plt.xlabel('Country',fontsize=15)
plt.ylabel('GDP Per Capita ($USD)',fontsize = 15)
plt.xticks(rotation = 45)
plt.savefig('/Users/isamarcortes/Dropbox/Isamar/Papers_In_Prep/Paper_1/Figures/GDP_PerCapita.png',dpi=600,bbox_inches='tight')
```
