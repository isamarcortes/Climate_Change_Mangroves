# SPEI Visualization
This code is used to visualize the Standardized Precipitation-Evapotranspiration Index. The data was gathered from the World Bank Climate Change Knowledge Portal in which they use the multimodel ensemble to calculate the SPEI.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
    
```

```python
def plot(past,future26,future45,future70,variableName,region):
    pastdata = pd.read_csv(past)
    future26_data = pd.read_csv(future26)
    future45_data = pd.read_csv(future45)
    future70_data = pd.read_csv(future70)
    
    plt.figure(figsize=(10, 3))
    plt.plot(pastdata['Time'],pastdata[variableName],label='Past')
    plt.plot(future26_data['Time'],future26_data[variableName],label='SSP 2.6')
    plt.plot(future45_data['Time'],future45_data[variableName],label='SSP 4.5')
    plt.plot(future70_data['Time'],future70_data[variableName],label='SSP 7.0')
    plt.legend()
    plt.title(region,fontsize=20)
    plt.ylabel('SPEI',fontsize=20)
    plt.xlabel('Year',fontsize=20)
    plt.xticks(fontsize = 16)
    plt.yticks(fontsize = 16)
    plt.savefig('/Users/isamarcortes/Dropbox/Isamar/Papers_In_Prep/Paper_1/Figures/SPEI_Figs/'+region+'.svg',format='svg',dpi=600)
    plt.show()
    
```
