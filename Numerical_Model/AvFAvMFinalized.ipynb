## Code works to create figure to compare field and model results

```python
###So is set to 35 ppt

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

PlanetSentinel = '/Users/isamarcortes/Dropbox/Isamar/Papers_In_Prep/Paper_1/SupplementaryMaterial/DatabaseFinalized.xlsx'
Data = pd.read_excel(PlanetSentinel)

sns.set_palette("colorblind")
#sns.set(rc={'axes.facecolor':'white'})

g= sns.scatterplot(x='LogAV_GEP',y='Log Vegetated Area Model',style='Island Name',hue='Island Name',data=Data)
g.legend_.set_title(None)
#m, b = np.polyfit(Data['LogAV_GEP'], Data['Log Vegetated Area Model'], 1)
x= np.linspace(3.7,6.6)
y=np.linspace(3.7,6.6)


#plt.plot(Data['LogAV_GEP'], m*Data['LogAV_GEP']+b)
plt.plot(x,y)

plt.xlabel('$A_{v}$ Field ($m^2$)',fontsize=18)
plt.xticks(fontsize=14)
plt.ylabel('$A_{v}$ Model ($m^2$)',fontsize=18)
plt.yticks(fontsize=14)
plt.title('Vegetated Area ($A_{v}$)',fontsize=20)



#plt.savefig('/Users/isamarcortes/Dropbox/Isamar/Papers_In_Prep/Paper_1/Figures/Planet_AvModelFeb26.png',dpi=600,bbox_inches='tight',format='png')
```
